## **Import The Necessary Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from lifelines import WeibullFitter, LogNormalFitter, LogLogisticFitter, KaplanMeierFitter
from lifelines import WeibullAFTFitter, LogNormalAFTFitter, LogLogisticAFTFitter
```

## **Get the Data**

```python
telco = pd.read_csv('telco.csv') 
telco.head()
```

## **Categorical Features**

```python
# Turning some of the features into categorical
cols = ['region','marital','ed','gender', 'retire','voice','internet','forward', 'custcat', 'churn']
 
categorical_df = pd.get_dummies(telco,
               columns=cols,
               prefix=cols,
               drop_first=False)
 
categorical_df.head()
```

```python
categorical_df.columns
```

## **DataFrame TO GO**

## Data Preprocessing for Categorical Variables

During the data preprocessing phase, certain columns in the dataset needed to be treated as categorical variables. To achieve this, these columns were converted into categorical data types, resulting in the addition of dummy columns to the dataframe. Each dummy column represents a level within the categorical variable.

To facilitate further analysis and avoid multicollinearity issues, one level from each categorical variable was selected as a reference. This reference level serves as a baseline for measuring the effects of the other levels within the variable.

### Example:

Consider the categorical variable 'gender' with levels 'Male' and 'Female.' To conduct the analysis, only one level was retained in the dataframe, let's say 'gender_Male,' while 'gender_Female' was omitted. This process was repeated for other categorical variables with multiple levels.



```python
df = categorical_df[[ 'tenure', 'age', 'income', 
                     'region_Zone 2', 'region_Zone 3',
                     'marital_Married', 'churn_Yes',
                     'ed_Did not complete high school', 'ed_High school degree',
                     'ed_Post-undergraduate degree', 'ed_Some college',
                     'voice_Yes', 'retire_Yes', 'gender_Male',
                    'internet_Yes', 'forward_Yes',
                    'custcat_E-service', 'custcat_Plus service', 'custcat_Total service']]
df.head()
```

```python
df.info()
```

## **Fitters**

```python
wb_fitter = WeibullFitter()
log_norm_fitter = LogNormalFitter()
log_lgstc_fitter = LogLogisticFitter()
```

## **AFT Fitters**

### Weibull AFT Fitter

```python
wb_aft = WeibullAFTFitter()
wb_aft.fit(df, duration_col='tenure', event_col='churn_Yes')
wb_aft_pred = wb_aft.predict_survival_function(df).transpose()
wb_avg = wb_aft_pred.mean().to_list()
wb_aft_pred.head()
```

### Log Normal AFT Fitter

```python
log_norm_aft = LogNormalAFTFitter()
log_norm_aft.fit(df, duration_col='tenure', event_col='churn_Yes')
log_norm_aft_pred = log_norm_aft.predict_survival_function(df).transpose()
log_norm_avg = log_norm_aft_pred.mean().to_list()
log_norm_aft_pred.head()
```

### Log Logistic AFT Fitter


```python
log_lgstc_aft = LogLogisticAFTFitter()
log_lgstc_aft.fit(df, duration_col='tenure', event_col='churn_Yes')
log_lgstc_aft_pred = log_lgstc_aft.predict_survival_function(df).transpose()
log_lgstc_avg = log_lgstc_aft_pred.mean().to_list() #computing the average survival time for each month
log_lgstc_aft_pred.head()
```

## **Plot of the 3 Methods**

```python
plt.plot(wb_avg, color = "red", label = "Weibull")
plt.plot(log_norm_avg, color = "blue", label = "Log-Normal")
plt.plot(log_lgstc_avg, color = 'green', label = 'Log-Logistic')
plt.legend()
plt.show()
```

## **information Criterions**

### AIC

```python
for i in [wb_fitter, log_norm_fitter, log_lgstc_fitter]:
    i.fit(durations = df["tenure"], event_observed = df["churn_Yes"])
    print(f"AIC {i.__class__.__name__}: {i.AIC_}")
```

The Akaike Information Criterion (AIC) is a measure of the relative quality of statistical models. A lower AIC value indicates a better-fitting model, considering both the goodness of fit and the simplicity of the model. 

The Log-Normal has the lowest AIC value (3209.04), suggesting that it provides the best trade-off between model fit and complexity among the three models considered.

Therefore, based on the AIC values alone, the Log-Normal model is preferred over the Weibull and Log-Logistic models. It's important to note that model selection should not solely rely on AIC; we should also consider the assumptions of the chosen distribution and validate the model's performance on independent datasets if possible.

### BIC

```python
for i in [wb_fitter, log_norm_fitter, log_lgstc_fitter]:
    i.fit(durations = df["tenure"], event_observed = df["churn_Yes"])
    print(f"BIC {i.__class__.__name__}: {i.BIC_}")
```

The Bayesian Information Criterion (BIC) is another criterion for model selection, similar to AIC. It penalizes models for complexity more strongly than AIC. Lower BIC values indicate a better trade-off between goodness of fit and model complexity. Based on the BIC values you provided:

Again, the Log-Normal has the lowest BIC value (3218.85), suggesting that it provides the best balance between model fit and complexity among the three models.

Consistent with the AIC results, the Log-Normal model appears to be the preferred choice based on the BIC values. It's recommended to consider both AIC and BIC, as well as the assumptions of the chosen distribution, when making a decision about the best-fitting model. Additionally, cross-validation or other model validation techniques could further support the model selection process.

## **CLV**

**Note:** We can calculate CLV by assuming that discount rate (r) is 10% and average monthly margin is 1300 AMD.

```python
MM = 1300
sequence = range(1,len(log_norm_aft_pred.columns)+1)
r = 0.1

for num in sequence:
    log_norm_aft_pred.loc[:,num] = log_norm_aft_pred.loc[:,num]/((1+r/12)**(sequence[num-1]-1))

log_norm_aft_pred.head()
```

```python
log_norm_aft_pred["CLV"] = MM * log_norm_aft_pred.sum(axis = 1)
log_norm_aft_pred.head()
```

```python
log_norm_aft_pred.describe()
```

```python
telco["CLV"] = log_norm_aft_pred["CLV"]
telco.head()
```

## **Assuming the data represents the population, how much would be your annual retention budget?**
### (Report (3))

```python
# Step 1: Calculate CLV
# Done

# Convert 'churn' column to numeric (Yes' means churned and 'No' means not churned)
telco['churn'] = telco['churn'].map({'Yes': 1, 'No': 0})

# Step 2: Determine Survival Probabilities using Kaplan-Meier estimator
kmf = KaplanMeierFitter()
kmf.fit(durations=telco['tenure'], event_observed=telco['churn'])
survival_probabilities = kmf.survival_function_at_times(telco['tenure']).reset_index()
survival_probabilities.columns = ['tenure', 'survival_probability']

# Step 3: Merge Survival Probabilities with the original DataFrame
telco = pd.merge(telco, survival_probabilities, on='tenure')

# Step 4: Identify At-Risk Subscribers
at_risk_threshold = 0.5  # Arbitrary threshold based on business needs
at_risk_subscribers = telco[telco['churn'] > at_risk_threshold]

# Step 5: Estimate Retention Budget
total_retention_budget = (at_risk_subscribers['CLV'] * at_risk_subscribers['survival_probability']).sum()

print(f"Estimated Annual Retention Budget: ֏{total_retention_budget:.2f}")
```
